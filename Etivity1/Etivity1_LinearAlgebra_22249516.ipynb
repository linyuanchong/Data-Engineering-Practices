```python
vectorOne = [20, 45, 12]
vectorTwo = [12, 71, 30]
vectorThree = [34, 56, 90]
vectorFour = [23, 69, 45]
vectorFive = [20, 12]
vectorSix = [12, 30]
vectorSeven = [34, 90]
vectorEight = [23, 45]

#[ 8 4 2 ]
#[ 4 7 6 ]
matrixOne = [[8, 4, 2], [4, 7, 6]]
#[ 8 2 ]
#[ 4 6 ]
matrixTwo = [[4, 7], [2, 6]]
#[ 6 9 ]
#[ 2 4 ]
matrixThree = [[6, 9], [2, 4]]
#[ 4 4 ]
#[ 8 8 ]
matrixFour = [[4, 4], [8, 8]]

scale = 2



#1) Scalar multiplication of a vector.
def scalarMultiplication(vector, scale):
  
  vecLen = len(vector)
  thisVec = []

  for x in range(vecLen):
    thisVec.append(vector[x]*scale)

  print("1) SCALAR MULTIPLICATION:")
  print("Solution:\t" + str(vector) + " * " + str(scale) + "\n\t\t= " + str(thisVec))
  print("\n")
  return thisVec



#2) Addition and subtraction of two vectors.
def addAndSub(firstVec, secVec):

  firstVecLen = len(firstVec)
  secVecLen = len(secVec)
  addVec = []
  subVec = []

  if firstVecLen == secVecLen:
    for x in range(secVecLen):
      addVec.append((firstVec[x] + secVec[x]))
      subVec.append((firstVec[x] - secVec[x]))
    print("2.1) ADDITION:")
    print("Solution:\t" + str(firstVec) + " + " + str(secVec) + "\n\t\t= " + str(addVec))
    print("2.2) SUBTRACTION:")
    print("Solution:\t" + str(firstVec) + " - " + str(secVec) + "\n\t\t= " + str(subVec))
    print("\n")

  elif firstVecLen != secVecLen:
    print("Length of both vectors do not match.")
    print("\n")

  return addVec, subVec



#3) The dot product of two vectors of length n.
def dotProd(firstVec, secVec):

  firstVecLen = len(firstVec)
  secVecLen = len(secVec)
  tempSum = 0
  totalProd = 0

  print("3) DOT PRODUCT:")

  if firstVecLen == secVecLen:
    for x in range(secVecLen):
      tempSum = firstVec[x]*secVec[x]
      print("Value " + str(x+1) + ":\t" + str(firstVec[x]) + " * " + str(secVec[x]) + " = " + str(tempSum))
      totalProd += tempSum
    print("Dot product:\t" + str(totalProd))
    print("\n")

  elif firstVecLen != secVecLen:
    print("Length of both vectors do not match.")
    print("\n")

  return totalProd



#4) Multiply a vector with a matrix of the appropriate size.
def vecMatMul(vector, matrix):

  vecLen = len(vector)
  matCol = len(matrix[0])
  matRow = len(matrix)
  newVec = []
  store = 0

  print("4) VECTOR AND MATRIX MULTIPLICATION:")

  if vecLen == matCol:
    print("Vector:\t\t" + str(vector))
    print("Matrix:\t\t" + str(matrix))
    print("Vector size:\t" + str(vecLen))
    print("Matrix column:\t" + str(matCol))
    print("Matrix row:\t" + str(matRow))

    for x in range(matRow):
      for y in range(matCol):
        store += (matrix[x][y])*(vector[y])
      newVec.append(round(store, 2))
      store = 0

    print("Product:\t" + str(newVec))
    print("\n")

  elif vecLen != matCol:
    print("Vector size does not match matrix columns.")
    print("\n")

  return newVec



#5) Calculate the inverse of a 2x2 matrix.
def inverseMat(matrix):

  matCol = len(matrix[0])
  matRow = len(matrix)
  det = (matrix[0][0]*matrix[1][1]) - (matrix[0][1]*matrix[1][0])
  con = 1/det
  newMat = [[(matrix[1][1]), -(matrix[0][1])], [-(matrix[1][0]),(matrix[0][0])]]
  invMat = [[], []]

  print("5) INVERSE MATRIX:")
  print("Determinant:\t" + str(det))

  if matCol == 2 and matRow == 2:
    print("2x2 matrix:\tTrue")

    if det != 0:
      print("Invertible:\tTrue")

      for x in range(matRow):
        for y in range(matCol):
          invMat[x].append(round(((newMat[x][y])*con), 2))

      print("Solution:\t" + str(matrix) + "^-1")
      print("\t\t= " + "1/" + str(det) + "\t" + str(newMat))
      print("\t\t= " + str(con) + "\t" + str(newMat))
      print("\t\t= " + str(invMat))
      print("\n")

    elif det == 0:
      print("Invertible:\tFalse")   
      print("\n")

  elif matCol != 2 or matRow != 2:
    print("2x2 matrix:\tFalse.")
    print("\n")

  return invMat

#Function for matrix multiplication (2x2 matrices only).
def matMul(firstVec, secVec):

  print("2x2 MATRIX MULTIPLICATION:")
  result = []
  a = firstVec[0][0]
  b = firstVec[0][1]
  c = firstVec[1][0]
  d = firstVec[1][1]
  e = secVec[0][0]
  f = secVec[0][1]
  g = secVec[1][0]
  h = secVec[1][1]

  result = [[round((a*e)+(b*g), 1), round((a*f)+(b*h), 1)], [round((c*e)+(d*g), 1), round((c*f)+(d*h), 1)]]

  print("Product of " + str(firstVec) + " * " + str(secVec) + " = " + str(result))

  return result
```

Unlike humans, machines are capable of storing and studying an enormous amount of data due to the possession of non-volatile memory. A huge amount of data is fed to machines in order to allow them to iteratively process and study the data, then slowly learn by analogy. However, humans are incapable of perfectly filtering out incompatible data before feeding them to machines due to intelligence restrictions. Machines themselves are expected to have the ability to consider and review the data provided, and this is where one of the biggest uses of Linear Algebra in Machine Learning comes into place. Machines are able to neglect irrelevant data values such as outliers based on how the data is plotted on a graph. Data modeled after each iteration might also come in bad shapes such as overfitting and underfitting. Linear Algebra helps in evaluating data models produced from each iteration. Overfitting describes a data model that fits the existing observations too well but fails to predict future new observations (Liu, 2020). In this case, with the help of Linear Algebra, researchers are able to evaluate data models and make further decisions from there.

A good data model is capable of a lot of uses, and providing recommendations based on it is one of them. With an appropriate amount of data and correct data modeling done, a data model will show a pattern and tell how consistent the data values are. Linear Algebra is capable of helping researchers discover specific patterns from these data values. These patterns can later be used to construct various functionalities. One of the most typical examples of this is a recommendation system. Recommendation systems are usually embedded as a passive feature in many applications these days, users’ activity data are collected and processed to be provided with more relevant content that can inflict larger interests (Brownlee, 2018).

Linear Algebra also aids in deep learning. In fact, Linear Algebra is capable of describing deep learning methods using matrix notations. In order to execute neural networks, Linear Algebra is needed to structurize data in order for the data to be eligible for processing.
